```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

class Encoder:
    def __init__(self, df):
        self.df = df
        self.encoded_df = df.copy()
        
    def one_hot_encode(self, columns_to_encode):
        one_hot_encoded = pd.get_dummies(self.encoded_df[columns_to_encode])
        self.encoded_df = pd.concat([self.encoded_df, one_hot_encoded], axis=1)
        self.encoded_df = self.encoded_df.drop(columns_to_encode, axis=1)
        
    def label_encode(self, columns_to_encode):
        for col in columns_to_encode:
            le = LabelEncoder()
            self.encoded_df[col] = le.fit_transform(self.encoded_df[col])
        
    def unique_counts(self, ignore_columns=[]):
        p = [col for col in self.df.columns if col not in ignore_columns and self.df[col].dtype == type(object) and self.df[col].nunique() <= 10]
        q = [col for col in self.df.columns if col not in ignore_columns and self.df[col].dtype == type(object) and self.df[col].nunique() > 10]
        print("The Columns which having unique counts less than 10:")
        print(p)

        print("The Columns which having unique counts greater than 10:")
        print(q)
        return p, q
            
    def get_encoded_df(self):
        return self.encoded_df
    
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
#import seaborn as sns
from matplotlib.pyplot import figure

class Correlation_Class:

    def pearson_corr(self, df):
        corr_mat = df.corr(method='pearson')
        sorted_mat = corr_mat.unstack().sort_values(ascending=False).drop_duplicates()
        sorted_mat = sorted_mat[sorted_mat>=.7]
        print(sorted_mat)
        correlated_pairs = list(sorted_mat.index)
        return correlated_pairs

    def spearman_corr(self, df):
        return df.corr(method='spearman')

    def plot_heatmap(self, df):

        figure(figsize=(15, 15), dpi=80)

        mask = np.triu(np.ones_like(df.corr()))
        dataplot = sns.heatmap(df.corr(), cmap="YlGnBu", annot=True, mask=mask)

        #dataplot.set(axis_bgcolor='k')
        #plt.rc('font', **{'size': 10.0})

        #dataplot.savefig("output.png")
        plt.show()

    def Correlation_data(self, df):
        # Extract Significant Parameters having higher Correlation 
        #pd.set_option('display.max_rows', None)

        corr_abs = df.corr().abs()
        corr_abs_val = corr_abs.unstack()
        sorted_corr = corr_abs_val.sort_values(kind="quicksort", ascending=False)
        significant_corr = sorted_corr[(sorted_corr < 1) & (sorted_corr > np.mean(sorted_corr))]
        significant_corr_df = pd.DataFrame(significant_corr, columns = ['correlation'])
        significant_corr_df.reset_index(level=[0,1], inplace=True)
        significant_corr_df = significant_corr_df.rename(columns={'level_0': 'param_1', 'level_1': 'param_2'})

        significant_corr_df.drop_duplicates(subset=['correlation'], keep = 'last', inplace = True)
        significant_corr_df.reset_index(inplace=True)
        del significant_corr_df['index']

        print('Significant Correlation among Parameters')
        display(significant_corr_df)
        #sig_cor_90 = significant_corr_df[significant_corr_df.correlation >=.8]
        #sig_cor_90 = sig_cor_90['param_1'],sig_cor_90['param_2']
        #sig_cor_90 = list(sig_cor_90)
        #return sig_cor_90
        correlated_features = significant_corr_df[(significant_corr_df['correlation'] > 0.8)]
        correlated_features_list = list(correlated_features['param_1'].unique()) + list(correlated_features['param_2'].unique())

        return correlated_features_list
```

```python
# load statmodels functions
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant
```

```python
# compute the vif for all given features
def compute_vif(df,considered_features):
    
    
    X = df[considered_features]# the calculation of variance inflation requires a constant
    X['intercept'] = 1# create dataframe to store vif values
    
    vif = pd.DataFrame()
    
    vif["Variable"] = X.columns
    
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    
    vif = vif[vif['Variable']!='intercept']
    
    return vif


# compute vif 
compute_vif(scaled_df,scaled_df.columns).sort_values('VIF', ascending=False)
```

```python
def get_high_vif_features(df, considered_features, vif_threshold=10):
    X = df[considered_features]
    X['intercept'] = 1

    vif = pd.DataFrame()
    vif["Variable"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    vif = vif[vif['Variable']!='intercept']

    high_vif_features = list(vif[vif['VIF'] >= vif_threshold]['Variable'])
    inf_vif_features = list(vif[vif['VIF'] == np.inf]['Variable'])

    return high_vif_features + inf_vif_features
```
